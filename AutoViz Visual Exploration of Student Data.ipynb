```python
#  AutoViz Visual Exploration of Student Data

import sys
import subprocess
import pkg_resources

# Install AutoViz if missing
required = {'autoviz'}
installed = {pkg.key for pkg in pkg_resources.working_set}
missing = required - installed

if missing:
    python = sys.executable
    subprocess.check_call([python, '-m', 'pip', 'install', *missing], stdout=subprocess.DEVNULL)

# Imports
import pandas as pd
import numpy as np
from autoviz.AutoViz_Class import AutoViz_Class
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

# Load dataset
students = pd.read_csv("https://raw.githubusercontent.com/harmanani/AAI614/main/Week%202/students.csv")

# Set target variable
target_variable = 'Avg_Hours_Studied_per_week'

# Run AutoViz
AV = AutoViz_Class()
sep = ','

dft = AV.AutoViz('', sep=sep, depVar=target_variable, dfte=students,
                 header=0, verbose=1, lowess=False,
                 chart_format='bokeh', max_rows_analyzed=500, max_cols_analyzed=30)
```
