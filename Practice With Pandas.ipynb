```python
# Imports
import pandas as pd
import numpy as np
import ssl

# Fix SSL certificate issue
ssl._create_default_https_context = ssl._create_unverified_context

# Load NBA dataset
nba = pd.read_csv("https://raw.githubusercontent.com/harmanani/AAI614/main/Week%202/nba.csv")

print(nba)
# Basic Series dtype check
print(pd.Series([1, 2, 3]).dtype)

# DataFrame structure
print(nba.dtypes)
print(nba.dtypes.value_counts())
print(nba.index)
print(nba.columns)
print(nba.ndim)
print(nba.shape)
print(nba.size)
print(nba.count())
print(nba.count().sum())

# Handling missing data
data = {
    "A": [1, np.nan],
    "B": [2, 3]
}
df = pd.DataFrame(data)
print(df)
print(df.size)
print(df.count())
print(df.count().sum())

# Shared methods between Series and DataFrames
print(nba.head(2))
print(nba.tail(3))
print(nba.sample(3))
print(nba.nunique())
print(nba.max(numeric_only=True))
print(nba.min(numeric_only=True))
print(nba.nlargest(4, "Salary"))

# Convert Birthday to datetime
nba["Birthday"] = pd.to_datetime(nba["Birthday"])
print(nba.nsmallest(3, "Birthday"))

# Summary statistics
print(nba.sum(numeric_only=True))
print(nba.mean(numeric_only=True))
print(nba.median(numeric_only=True))
print(nba.mode(numeric_only=True))
print(nba.std(numeric_only=True))

# Sorting
print(nba.sort_values("Name").head())
print(nba.sort_values("Name", ascending=False).head())
print(nba.sort_values("Birthday", ascending=False).head())
print(nba.sort_values(["Team", "Name"]).head())
print(nba.sort_values(["Team", "Salary"], ascending=[True, False]).head())

# Sorting by index
nba = nba.sort_index()
print(nba.sort_index(ascending=False).head())
print(nba.sort_index(axis=1).head())
print(nba.sort_index(axis=1, ascending=False).head())

# Setting index
nba = nba.set_index("Name")

# Selecting columns
print(nba["Salary"].head())
print(nba["Position"].head())
print(nba[["Salary", "Birthday"]].head())
print(nba.select_dtypes(include="object").head())
print(nba.select_dtypes(exclude=["object", "int"]).head())

# Selecting rows by label
print(nba.loc["LeBron James"])
print(nba.loc[["Kawhi Leonard", "Paul George"]])
print(nba.sort_index().loc["Otto Porter":"Patrick Beverley"])
print(nba.sort_index().loc["Zach Collins":])
print(nba.sort_index().loc[:"Al Horford"])

# Selecting rows by position
print(nba.iloc[300])
print(nba.iloc[[100, 200, 300, 400]])
print(nba.iloc[400:404])
print(nba.iloc[:2])
print(nba.iloc[-10:-6])
print(nba.iloc[0:10:2])

# Extracting specific values
print(nba["Team"].loc["Giannis Antetokounmpo"])
print(nba.loc["James Harden", ["Position", "Birthday"]])
print(nba.loc[["Russell Westbrook", "Anthony Davis"], ["Team", "Salary"]])
print(nba.loc["Joel Embiid", "Position":"Salary"])
print(nba.iloc[57, 3])
print(nba.iloc[100:104, :3])
print(nba.at["Austin Rivers", "Birthday"])
print(nba.iat[263, 1])

# Extracting from Series
print(nba["Salary"].loc["Damian Lillard"])
print(nba["Salary"].at["Damian Lillard"])
print(nba["Salary"].iloc[234])
print(nba["Salary"].iat[234])

# Renaming columns and rows
nba.columns = ["Team", "Position", "Date of Birth", "Pay"]
nba = nba.rename(columns={"Date of Birth": "Birthday"})
nba = nba.rename(index={"Giannis Antetokounmpo": "Greek Freak"})
print(nba.loc["Greek Freak"])

# Resetting index
nba = nba.reset_index().set_index("Team")
print(nba.head())
```
