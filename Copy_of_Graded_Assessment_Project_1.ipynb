````python
from ucimlrepo import fetch_ucirepo

# fetch dataset
adult = fetch_ucirepo(id=2)

# data (as pandas dataframes)
X = adult.data.features
y = adult.data.targets

# metadata
print(adult.metadata)

# variable information
print(adult.variables)
````

````python
import pandas as pd

# Concatenate X and y to create a single DataFrame
adult_df = pd.concat([X, y], axis=1)

# Save the combined DataFrame to a CSV file named adult.csv
adult_df.to_csv('adult.csv', index=False)

print("Combined 'adult' dataset saved to adult.csv")
````

````python
X.to_csv('X.csv', index=False)
y.to_csv('y.csv', index=False)
print("Features DataFrame saved to X.csv")
print("Targets DataFrame saved to y.csv")
````
