```python

# ========= Adult (Census Income)  Data Exploration =========
# Q1: Columns & dtypes
# Q2: Missing values and representation
# Q3: Distributions of capital_gain and capital_loss
# Q4: Suggested categorical transformation and distribution tables

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

#How are missing values represented in this data? #########
#Cast missing values to np.nan, if necessary. ############
#Count the number of missing values in each column.#######
# --- Read & clean ---
file_path = Path('/content/adult.csv')
na_tokens = ['?', ' ?']  # map unknown tokens to NaN
df = pd.read_csv(file_path, na_values=na_tokens)
# Strip whitespace from string columns
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype(str).str.strip()

# Normalize column names: use underscores instead of hyphens
df.columns = [c.strip().replace('-', '_') for c in df.columns]

# --- Q1: Inspect columns and dtypes ---
print("Columns:", list(df.columns))
print("\nDtypes:\n", df.dtypes)

# --- Q2: Missing values ---
missing_counts = df.isna().sum()
missing_pct = (df.isna().mean() * 100).round(2)
missing_summary = pd.DataFrame({'missing_count': missing_counts,
                                'missing_pct': missing_pct})
print("\nMissing summary:\n", missing_summary)

# Q3: Distributions of capital_gain ########
plt.figure(figsize=(8,5))
plt.hist(df['capital_gain'].dropna(), bins=50, color='#4C78A8', edgecolor='white')
plt.title('Distribution of capital_gain')
plt.xlabel('capital_gain'); plt.ylabel('count'); plt.tight_layout()
plt.show()

#Q3: Plot the distributions of capital_loss #######
plt.figure(figsize=(8,5))
plt.hist(df['capital_loss'].dropna(), bins=50, color='#F58518', edgecolor='white')
plt.title('Distribution of capital_loss')
plt.xlabel('capital_loss'); plt.ylabel('count'); plt.tight_layout()
plt.show()

cg = df['capital_gain'].dropna()
cl = df['capital_loss'].dropna()
print(f"\nShare of zeros in capital_gain: {(cg==0).mean():.4f}")
print(f"Share of zeros in capital_loss: {(cl==0).mean():.4f}")

# --- Q4: Quantile binning for non-zero values + 'None' for zeros ---
def make_quantile_bins(series):
    q = series.quantile([0.33, 0.66])
    bins = [0, q.iloc[0], q.iloc[1], series.max()]
    labels = ['Low', 'Medium', 'High']
    return bins, labels

nonzero_cg = cg[cg > 0]
nonzero_cl = cl[cl > 0]

# capital_gain_cat
if len(nonzero_cg) > 0:
    bins_cg, labels_cg = make_quantile_bins(nonzero_cg)
    df['capital_gain_cat'] = pd.Series(np.nan, index=df.index, dtype='object') # Initialize with NaN as object type
    df.loc[df['capital_gain'] == 0, 'capital_gain_cat'] = 'None' # Assign 'None' for zeros
    df.loc[df['capital_gain'] > 0, 'capital_gain_cat'] = pd.cut(
        df.loc[df['capital_gain'] > 0, 'capital_gain'],
        bins=bins_cg, labels=labels_cg, include_lowest=True
    ).astype(str) # Convert categories to string
else:
    df['capital_gain_cat'] = 'None'

# capital_loss_cat
if len(nonzero_cl) > 0:
    bins_cl, labels_cl = make_quantile_bins(nonzero_cl)
    df['capital_loss_cat'] = pd.Series(np.nan, index=df.index, dtype='object') # Initialize with NaN as object type
    df.loc[df['capital_loss'] == 0, 'capital_loss_cat'] = 'None' # Assign 'None' for zeros
    df.loc[df['capital_loss'] > 0, 'capital_loss_cat'] = pd.cut(
        df.loc[df['capital_loss'] > 0, 'capital_loss'],
        bins=bins_cl, labels=labels_cl, include_lowest=True
    ).astype(str) # Convert categories to string
else:
    df['capital_loss_cat'] = 'None'

# Distributions of new categorical variables
cg_cat = pd.concat([
    df['capital_gain_cat'].value_counts(dropna=False).to_frame('count'),
    (df['capital_gain_cat'].value_counts(normalize=True, dropna=False)*100).round(2).to_frame('pct')
], axis=1)
cl_cat = pd.concat([
    df['capital_loss_cat'].value_counts(dropna=False).to_frame('count'),
    (df['capital_loss_cat'].value_counts(normalize=True, dropna=False)*100).round(2).to_frame('pct')
], axis=1)

print("\ncapital_gain_cat distribution:\n", cg_cat)
print("\ncapital_loss_cat distribution:\n", cl_cat)
```

```python

# Step by Step: LOAD + CLEAN + PLOTS + OUTLIERS + REGRESSION
#  Explore the distribution of fnlwgt.

import pandas as pd, numpy as np, matplotlib.pyplot as plt
import statsmodels.api as sm

# Load and normalize columns

df = pd.read_csv('/content/adult.csv', na_values=['?', ' ?'])
df.columns = [c.strip().replace('-', '_') for c in df.columns]
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype(str).str.strip()

# ----- Explore fnlwgt -----
fn = df['fnlwgt']

# Overall histogram
plt.figure(figsize=(8,5))
plt.hist(fn.dropna(), bins=60, color='#4C78A8', edgecolor='white')
plt.title('Distribution of fnlwgt (overall)')
plt.xlabel('fnlwgt'); plt.ylabel('count'); plt.tight_layout()
plt.savefig('fnlwgt_hist_overall.png'); plt.close()

# By sex histograms
#Compare the distribution of this variable between men and women and comment on any trends you notice.
fig, axes = plt.subplots(1, 2, figsize=(12,4), sharey=True)
for ax, sex_val, color in zip(axes, ['Male','Female'], ['#1F77B4','#FF7F0E']):
    data = df.loc[df['sex']==sex_val, 'fnlwgt'].dropna()
    ax.hist(data, bins=60, color=color, edgecolor='white')
    ax.set_title(f'fnlwgt distribution ({sex_val})'); ax.set_xlabel('fnlwgt')
axes[0].set_ylabel('count'); plt.tight_layout()
plt.savefig('fnlwgt_hist_by_sex.png'); plt.close()

# Boxplots by sex
plt.figure(figsize=(6,5))
plt.boxplot([df.loc[df['sex']=='Male','fnlwgt'].dropna(),
             df.loc[df['sex']=='Female','fnlwgt'].dropna()],
            tick_labels=['Male','Female'], showfliers=True)
plt.title('fnlwgt boxplots by sex'); plt.ylabel('fnlwgt')
plt.tight_layout(); plt.savefig('fnlwgt_box_by_sex.png'); plt.close()

# Summary tables
summary_overall = fn.describe(percentiles=[0.25,0.5,0.75]).to_frame('overall')
summary_overall.loc['skew'] = fn.skew()
summary_overall.loc['kurtosis_excess'] = fn.kurt()  # excess kurtosis
summary_overall.to_csv('fnlwgt_summary_overall.csv')

by_sex = df.groupby('sex')['fnlwgt'].agg(['count','mean','std','min','median','max'])
by_sex['q1'] = df.groupby('sex')['fnlwgt'].quantile(0.25)
by_sex['q3'] = df.groupby('sex')['fnlwgt'].quantile(0.75)
by_sex['skew'] = df.groupby('sex')['fnlwgt'].apply(lambda s: s.skew())
by_sex['kurtosis_excess'] = df.groupby('sex')['fnlwgt'].apply(lambda s: s.kurt())
by_sex.to_csv('fnlwgt_summary_by_sex.csv')

# ----- Outliers: IQR rule (if excluding, set to missing) -----
Q1, Q3 = fn.quantile(0.25), fn.quantile(0.75)
IQR = Q3 - Q1
lower_bound, upper_bound = Q1 - 1.5*IQR, Q3 + 1.5*IQR
outlier_mask = (df['fnlwgt'] < lower_bound) | (df['fnlwgt'] > upper_bound)
pd.DataFrame({
    'lower_bound':[lower_bound], 'upper_bound':[upper_bound],
    'num_outliers':[int(outlier_mask.sum())],
    'total_rows':[len(df)],
    'pct_outliers':[round(100*outlier_mask.sum()/len(df),2)]
}).to_csv('fnlwgt_outlier_info.csv', index=False)

# Set outlier fnlwgt to NaN for downstream (keep original for reference)
df['fnlwgt_original'] = df['fnlwgt']
df.loc[outlier_mask, 'fnlwgt'] = np.nan

# Plot (after outlier removal)
plt.figure(figsize=(8,5))
plt.hist(df['fnlwgt'].dropna(), bins=60, color='#2CA02C', edgecolor='white')
plt.title('fnlwgt distribution (after IQR outlier removal)')
plt.xlabel('fnlwgt'); plt.ylabel('count'); plt.tight_layout()
plt.savefig('fnlwgt_hist_after_outliers.png'); plt.close()

# ----- Regression models -----
# Binary variables
#Fit a linear regression with hours_per_week as the dependent variable and sex as the independent variable.
df['sex_binary'] = df['sex'].map({'Male':1, 'Female':0})
df['gross_income_group'] = df['income'].map({'<=50K':0,'>50K':1,'<=50K.':0,'>50K.':1})

def fit_ols(cols):
    data = df[cols].dropna()
    y = data['hours_per_week']
    X = sm.add_constant(data.drop(columns=['hours_per_week']))
    return sm.OLS(y, X).fit()

m1 = fit_ols(['hours_per_week','sex_binary'])
m2 = fit_ols(['hours_per_week','sex_binary','education_num'])
m3 = fit_ols(['hours_per_week','sex_binary','education_num','gross_income_group'])

# Save model summaries
open('model1_summary.txt','w').write(m1.summary().as_text())
open('model2_summary.txt','w').write(m2.summary().as_text())
open('model3_summary.txt','w').write(m3.summary().as_text())

# Comparison tables
pd.DataFrame({
    'model':['M1: hours ~ sex','M2: + education_num','M3: + gross_income_group'],
    'n_obs':[int(m1.nobs),int(m2.nobs),int(m3.nobs)],
    'R2':[m1.rsquared,m2.rsquared,m3.rsquared],
    'Adj_R2':[m1.rsquared_adj,m2.rsquared_adj,m3.rsquared_adj],
    'AIC':[m1.aic,m2.aic,m3.aic], 'BIC':[m1.bic,m2.bic,m3.bic]
}).to_csv('models_metrics_consistent.csv', index=False)

coef_rows = []
for name, model in [('M1',m1),('M2',m2),('M3',m3)]:
    coefs = model.params
    pvals = model.pvalues
    for term in coefs.index:
        coef_rows.append({'model':name,'term':term,'coef':coefs[term],'pvalue':pvals[term]})
pd.DataFrame(coef_rows).to_csv('models_coefficients_consistent.csv', index=False)

# Boxplot: hours_per_week by sex
plt.figure(figsize=(6,5))
plt.boxplot([df.loc[df['sex']=='Male','hours_per_week'].dropna(),
             df.loc[df['sex']=='Female','hours_per_week'].dropna()],
            tick_labels=['Male','Female'], showfliers=True)
plt.title('Hours per week by sex'); plt.ylabel('hours_per_week')
plt.tight_layout(); plt.savefig('hours_box_by_sex.png'); plt.close()

# Quick numeric table
df.groupby('sex')['hours_per_week'].agg(['count','mean','median','std','min','max']).to_csv('hours_per_week_by_sex.csv')
```

```python
#correlation

# ========= Correlation & Covariance Analysis for Adult (Census Income) =========
# Computes:
# 1) Pearson correlations among age, education_num, hours_per_week
# 2) Two-sided significance tests for correlations (p-values)
# 3) Male vs Female comparison: corr(age, education_num)
# 4) Covariance matrix for education_num and hours_per_week
# Saves results to CSV files for reproducibility.

import pandas as pd
import numpy as np

# Try to import scipy for exact p-values; fall back to an approximation if unavailable.
try:
    from scipy import stats
    HAVE_SCIPY = True
except Exception:
    HAVE_SCIPY = False
    import math

# ---------- Load & clean ----------
df = pd.read_csv('/content/adult.csv', na_values=['?', ' ?'])

# Normalize column names: replace hyphens with underscores, strip whitespace
df.columns = [c.strip().replace('-', '_') for c in df.columns]

# Strip whitespace from string/object columns
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype(str).str.strip()

# Keep only relevant columns; drop rows with missing values in these columns
cols = ['age', 'education_num', 'hours_per_week', 'sex']
d = df[cols].dropna()

# ---------- 1) Pearson correlation matrix ----------
corr_mat = d[['age', 'education_num', 'hours_per_week']].corr(method='pearson')
corr_mat.to_csv('correlation_matrix_age_edu_hours.csv')  # export

print("Correlation matrix:")
print(corr_mat.round(4))
print()

# ---------- 2) Pairwise correlations + p-values ----------
pairs = [
    ('age', 'education_num'),
    ('age', 'hours_per_week'),
    ('education_num', 'hours_per_week')
]

rows = []
for a, b in pairs:
    x = d[a].astype(float).values
    y = d[b].astype(float).values
    n = len(x)

    if HAVE_SCIPY:
        r, p = stats.pearsonr(x, y)
    else:
        # Pearson r via numpy; approximate p-value using t-statistic and Normal approx if SciPy unavailable
        r = np.corrcoef(x, y)[0, 1]
        t = r * ((n - 2) ** 0.5) / ((1 - r ** 2) ** 0.5)
        # Approximate two-sided p-value using Normal survival (less accurate than Student's t)
        p = 2 * (1 - 0.5 * (1 + math.erf(abs(t) / np.sqrt(2))))

    rows.append({'var_x': a, 'var_y': b, 'r': r, 'p_value': p, 'n': n})

pairwise_df = pd.DataFrame(rows)
pairwise_df.to_csv('pairwise_correlations_pvalues.csv', index=False)

print("Pairwise correlations with p-values:")
print(pairwise_df.round(4))
print()

# Flag correlations with |r| > 0.1
flag_df = pairwise_df[pairwise_df['r'].abs() > 0.1]
flag_df.to_csv('correlations_gt_0_1.csv', index=False)

if not flag_df.empty:
    print("Flagged pairs (|r| > 0.1):")
    print(flag_df.round(4))
else:
    print("No pairs exceed |r| > 0.1.")
print()

# ---------- 3) Male vs Female: corr(age, education_num) ----------
male = d[d['sex'] == 'Male'][['age', 'education_num']].dropna()
female = d[d['sex'] == 'Female'][['age', 'education_num']].dropna()

def corr_with_p(df2):
    x = df2['age'].astype(float).values
    y = df2['education_num'].astype(float).values
    n = len(x)
    if HAVE_SCIPY:
        r, p = stats.pearsonr(x, y)
        return r, p, n
    else:
        r = np.corrcoef(x, y)[0, 1]
        t = r * ((n - 2) ** 0.5) / ((1 - r ** 2) ** 0.5)
        p = 2 * (1 - 0.5 * (1 + math.erf(abs(t) / np.sqrt(2))))
        return r, p, n

rm, pm, nm = corr_with_p(male)
rf, pf, nf = corr_with_p(female)

male_female = pd.DataFrame([
    {'group': 'Male', 'r_age_education_num': rm, 'p_value': pm, 'n': nm},
    {'group': 'Female', 'r_age_education_num': rf, 'p_value': pf, 'n': nf},
])
male_female.to_csv('correlation_age_edu_by_sex.csv', index=False)

print("Correlation(age, education_num) by sex:")
print(male_female.round(4))
print()

# ---------- 4) Covariance matrix (sample, N-1) ----------
cov_mat = d[['education_num', 'hours_per_week']].cov()
cov_mat.to_csv('covariance_education_hours.csv')

print("Covariance matrix for education_num and hours_per_week:")
print(cov_mat.round(4))
print()

# ---------- Optional: quick narrative prints ----------
# You can keep or remove these as needed.
r_edu_hours = pairwise_df.loc[
    (pairwise_df['var_x'] == 'education_num') & (pairwise_df['var_y'] == 'hours_per_week'), 'r'
].iloc[0]
p_edu_hours = pairwise_df.loc[
    (pairwise_df['var_x'] == 'education_num') & (pairwise_df['var_y'] == 'hours_per_week'), 'p_value'
].iloc[0]

print(f"education_num  hours_per_week: r = {r_edu_hours:.4f}, p-value = {p_edu_hours:.2e}")
print(f"Male r(age, education_num) = {rm:.4f}, p = {pm:.2e} (n={nm})")
print(f"Female r(age, education_num) = {rf:.4f}, p = {pf:.2e} (n={nf})")
```
